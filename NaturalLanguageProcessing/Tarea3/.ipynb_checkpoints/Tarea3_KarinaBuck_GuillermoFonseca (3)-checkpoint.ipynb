## Pregunta 1

### Adquisicion de base de datos

Comenzamos guardando en nuestra base de datos noticias recopiladas desde gamba.cl y el ciudadano.com, se guarda el archivo csv en 'github'y se adquieren los datos desde ese repositorio.

```python
import requests

dwn_url = 'https://raw.githubusercontent.com/memo26167/redes_neuronales_ufro/master/NaturalLanguageProcessing/Tarea3/articulos.csv'
url = requests.get(dwn_url)
url.encoding = 'UTF-8'
text = url.text
text = text.replace(u'\xa0', u' ')
```

### Transformacion de texto
Se transforma el texto adquirido de la base de datos, a un dataframe de pandas, utilizamos la librer√≠a "io" para hacerlo pasar por un "archivo". Con un total de 50 noticias relacionadas con los t√≥picos "Pi√±era" y "AFP", comenzamos a trabajar.

```python
from io import StringIO
csv_raw = StringIO(url.text)

import pandas as pd
df_documents = pd.read_csv(csv_raw, header=None)
df_documents
```

### Encodificacion y tokenizacion
Creamos el diccionario del corpus tokenizando cada uno de los 50 documentos que se posee el dataframe del corpus, descargamos nuestro tokenizador y las stop-words de la librer√≠a nltk, para tokenizar nuestro "mensaje".



```python
dict_corpus = {}

#from nltk.tokenize.casual import casual_tokenize
import re

from nltk.tokenize import TweetTokenizer
tweet_tokenizer = TweetTokenizer(strip_handles=True, reduce_len=True)

import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('spanish')

puntuacion = set((',', '.', '--', '-', '¬°', '!', '¬ø', '?', ':', ';', '``', "''", '(', ')', '[', ']', '¬´', '¬ª', '/', '‚Äî', '_', '..', '+', '‚Ä¶', '‚Äò', '‚Äô', '‚Äì', '%', '‚Äú', '‚Äù', '‚Ä≥', '"', '¬∑', '|', '<', '>', '=', '*', '¬∞'))

def tokenizacion(message):
  '''Devuelve un diccionario con el mensaje tokenizado 
     convirtiendo el mensaje a minusculas, utilizando TweetTokenizer 
     (ya que la base de datos posee tweets), eliminando stopwords y puntuacion
     para medir mejor la similitud
     El mensaje a recibir debe ser un string'''
  # Usamos tweet tokenizer
  # Ademas convertimos todo a minusculas
  tokens = tweet_tokenizer.tokenize(message.lower())

  # Eliminamos stopwords
  tokens = [x for x in tokens if x not in stopwords]
  # Eliminamos puntuacion
  tokens = [x for x in tokens if x not in puntuacion]
  return tokens


for i, sent in enumerate(df_documents.values):
  tokens = tokenizacion(sent[0])
  dict_corpus[f'sent{i}'] = dict((tok, 1) for tok in tokens)

dict_corpus['sent35']
```

### Bolsa de palabras
Se crear√° el dataframe de Bolsa de Palabras

```python
df = pd.DataFrame.from_records(dict_corpus).fillna(0).astype(int).T
print(df)
```

### Medicion de similitud
Ahora se medira la similitud entre los documentos del corpus:

1.   Creamos la variable similitud haciendo el producto punto en una lista por comprensi√≥n que se recorren dentro de nuestro corpus.
2.   Creamos la matriz similitud, donde es sim√©trica y su diagonal est√° compuesta por ceros.
3.   Para poder medir similitud entre textos, se escoge hacerlo a trav√©s de "aritmetica modular", ya que numpy para buscar el valor m√°ximo de una matriz, la convierte en array unidimensional y al buscar el valor m√°ximo, entrega la posici√≥n de ese array unidimensional.



```python
import numpy as np

similitud = [[doc1.dot(doc2) for doc2 in df.values] for doc1 in df.values]
similitud = np.array(similitud)
np.fill_diagonal(similitud, 0)

def entrega_max_similitud(similitud_matrix):
  max_index = similitud_matrix.argmax()
  n,_ = similitud_matrix.shape
  fila = max_index%n
  columna = max_index//n
  max_value = similitud_matrix[columna][fila]
  return max_value, columna, fila

max_value, columna, fila = entrega_max_similitud(similitud)
print("La similitud maxima es de", max_value, f"con los documentos {columna} y {fila}")
print("Los documentos m√°s similares son:")
print(df_documents.values[columna])
print(df_documents.values[fila])
```

## Pregunta 2

### Vectorizaci√≥n TF-IDF

Comenzamos la vectorizaci√≥n viendo el l√©xico para este corpus que contiene 50 elementos (noticias):

1.   Primero enlistamos los documentos y sus tokens
2.   definimos la variable de todos los tokens en los documentos
3.   Definimos el l√©xico, utulizamos la funci√≥n "sorted" para que nos devuelva el objeto en una lista ordenada
4.   Importamos la funcion "OrderedDict" de la librer√≠a "collections" para creaar un diccionario con los tokens y ceros.
5.   Creamos el vector td-idf como lista vac√≠a y comenzamos a recorrer cada noticia dentro del corpus, comenzando con la creaci√≥n de una matriz de ceros.
6.   Ingremos nuevamente a recorrer, esta vez recorremos cada elemento del diccionario.
7.   Finalmente se aplican las formulas correspondientes a TF e IDF

Tf(t,d)=f(t,d)/|L_d|

IDF(t,C)=|C|/|{ #d : t \in d}|


```python
docs = df_documents.values
doc_tokens = []
for doc in docs:
    doc_tokens += [sorted(tokenizacion(doc[0]))]

#todos los token de los docs
all_doc_tokens = sum(doc_tokens, [])

# L√©xico
lexicon = sorted(set(all_doc_tokens))

from collections import OrderedDict

zero_vector = OrderedDict((token, 0) for token in lexicon)
import copy
from collections import Counter
import math

tfidf_vectors = []
for doc in docs:
    vec = copy.copy(zero_vector)
    tokens = tokenizacion(doc[0])
    token_counts = Counter(tokens)
    for key, value in token_counts.items():
        containing_key = 0
        for _doc in docs:
            if key in _doc:
                containing_key += 1
        tf = value / len(lexicon)
        idf = len(docs) / (containing_key + 1)
        vec[key] = tf* math.log(idf)
    tfidf_vectors.append(vec)
```

## Calculo de similitud

Para el c√°lculo de similitud, comenzamos importando la librer√≠a math, que trae consigo la funci√≥n sim_coseno la c√∫al toma dos vectores y calcula su similitud. Creamos la lista por compresi√≥npara luego calcular elproducto punto.
Finalmente retoranmos la el % de similitud de los textos m√°s similiares.

```python
import math

def sim_coseno(vec1, vec2):
    vec1 = [val for val in vec1.values()]
    vec2 = [val for val in vec2.values()]
    
    dot_prod = 0
    for i, v in enumerate(vec1):
        dot_prod += v*vec2[i]
        
    norm_1 = math.sqrt(sum([x**2 for x in vec1]))
    norm_2 = math.sqrt(sum([x**2 for x in vec2]))
    
    return dot_prod / (norm_1 * norm_2)
```

```python
similitud_tfidf = [[sim_coseno(doc1, doc2) for doc2 in tfidf_vectors] for doc1 in tfidf_vectors]

similitud_tfidf = np.array(similitud_tfidf)
np.fill_diagonal(similitud_tfidf, 0)

max_value, columna, fila = entrega_max_similitud(similitud_tfidf)
print("La similitud maxima es de", max_value, f"con los documentos {columna} y {fila}")
print("Los documentos m√°s similares son:")
print(df_documents.values[columna])
print(df_documents.values[fila])
```
